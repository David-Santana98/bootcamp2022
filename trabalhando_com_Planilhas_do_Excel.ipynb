## ***TRABALHANDO COM PLANILHAS DO EXCEL***

```python
#IMPORTANDO A BIBLIOTECA
import pandas as pd
```

```python
#LEITURA DOS ARQUIVOS APÓS UPLOAD
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df1.head() #verificando as 5 linhas do conjunto de dados
```

```python
#JUNTANDO TODOS OS ARQUIVOS em 1 arquivo só
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Após concatenar os arquivos 
# exibir as 5 primeiras linhas
df.head()
```

```python
#Exibindo as 5 últimas linhas
df.tail()
```

```python
# verificando o tipo de dado de cada coluna
df.dtypes
```

```python
# Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes #verificando se o tipo de dado mudou
```

```python
df.sample(5) #pegando uma amostra(sample) do conjunto de dados
```

```python
df.head()
```

## ***TRATANDO VALORES FALTANTES***

```python
# CONSULTANDO LINHAS COM VALORES FALTANTES
df.isnull().sum()
```

```python
# SUBSTITUINDO OS VALORES NULOS PELA MÉDIA
df["Vendas"].fillna(df["Vendas"].mean(), inplace = True)
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
# Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace = True)
```

```python
# Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
# Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
# Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

## ***CRIANDO COLUNAS NOVAS***

```python
#Criando um coluna de receita 
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head() #retornando as 5 primeiras colunas agora com a coluna receita 
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"] #caso não tivessemos a Quantidade de produtos vendidos
```

```python
df.head() #retornando a quantidade de produtos vendidos
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita 
df["Receita"].min()
```

```python
#nlargest
df.nlargest(3, "Receita")
```

```python
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10) #Ordenando pela Receita
```

##***TRABALHANDO COM DATAS***

```python
#TRANSFORMANDO A COLUNA DE DATA EM TIPO INTEIRO
df["Data"] = df["Data"].view("int64") #.astype foi atualizado para .view(...)
```

```python
#Verificando o tipo de dado de cada coluna 
df.dtypes
```

```python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupando por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando um nova coluna com o ano
df["Ano_venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mês e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando data mais antiga 
df["Data"].min()
```

```python
#Calculando a diferença de dias 
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(10)
```

```python
# Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19.sample(20)
```
